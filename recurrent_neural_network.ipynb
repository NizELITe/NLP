```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD

# Step 1: Prepare the sentence and tokenize it
sentence = "I like to eat pizza"
words = sentence.lower().split()  # Tokenize and convert to lowercase

# Step 2: Create a co-occurrence matrix
window_size = 1  # Define the window size (1 means immediate neighbors)
vocab = sorted(set(words))  # Create a vocabulary (unique words in the sentence)
word_to_idx = {word: idx for idx, word in enumerate(vocab)}  # Map each word to an index
co_occurrence_matrix = np.zeros((len(vocab), len(vocab)))  # Initialize a square matrix

# Fill in the co-occurrence matrix based on the context window
for i, word in enumerate(words):
    word_idx = word_to_idx[word]
    # Look at neighbors within the window size
    for j in range(i - window_size, i + window_size + 1):
        if j >= 0 and j < len(words) and i != j:
            neighbor_idx = word_to_idx[words[j]]
            co_occurrence_matrix[word_idx, neighbor_idx] += 1

# Step 3: Convert co-occurrence matrix to a DataFrame for better readability
co_occurrence_df = pd.DataFrame(co_occurrence_matrix, columns=vocab, index=vocab)
print("Co-occurrence Matrix:\n", co_occurrence_df)

svd = TruncatedSVD(n_components=5)
word_embeddings = svd.fit_transform(co_occurrence_matrix)


embeddings_df = pd.DataFrame(word_embeddings, columns=["Dim1", "Dim2", "Dim3", "Dim4", "Dim5"], index=vocab)
print("\nWord Embeddings (5D):\n", embeddings_df)

word_embeddings_truncated = word_embeddings[:3, :]
print("\nTruncated Word Embeddings (3x5):\n", word_embeddings_truncated)
```

```python
import numpy as np
from collections import defaultdict

# Sample Data
texts = ["I love this product", "This is terrible", "Awesome!", "Waste of money"]
labels = [1, 0, 1, 0]  # 1 for positive, 0 for negative

# Tokenization and Vocabulary Creation
vocab = defaultdict(lambda: len(vocab))
processed_texts = [[vocab[word.lower()] for word in sentence.split()] for sentence in texts]
vocab = dict(vocab)

# Padding
max_length = max(len(sentence) for sentence in processed_texts)
padded_texts = [sentence + [0] * (max_length - len(sentence)) for sentence in processed_texts]


embedding_dim = U.shape[1]
embeddings = np.random.rand(len(vocab), embedding_dim)

U = np.array([[0.4, -0.3, 0.1, -0.2, 0.5],
              [0.1, -0.2, 0.3, 0.2, -0.4],
              [0.2, -0.1, 0.5, 0.4, -0.3]])

V = np.array([[0.2, 0.6, -0.1],
              [0.3, -0.2, 0.4],
              [-0.4, 0.1, 0.5]])

W = np.array([[0.2, -0.1, 0.3],
              [-0.1, 0.4, -0.2],
              [0.4, -0.3, 0.5]])

b = np.zeros(3)
c = np.zeros(3)
h_0 = np.array([0.1, -0.1, 0.2])


def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0, keepdims=True)


outputs = []
hidden_states = []

for sentence in padded_texts:
    h_t = h_0
    for word_idx in sentence:
        embed_i = embeddings[word_idx]


        a_t = b + np.dot(W, h_t) + np.dot(U, embed_i)


        h_t = np.tanh(a_t)


        o_t = c + np.dot(V, h_t)
        y_t = softmax(o_t)


        word = list(vocab.keys())[list(vocab.values()).index(word_idx)]
        outputs.append((word, y_t))
        hidden_states.append((word, h_t))


print("Outputs:")
for word, output in outputs:
    print(f"Output for word '{word}': {output}")

print("\nHidden States:")
for word, hidden_state in hidden_states:
    print(f"Hidden state for word '{word}': {hidden_state}")
```
