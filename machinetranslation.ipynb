**One Hot encoding with simple encoder decoder**

```python
import numpy as np


# Define characters from the processed sentences
english_characters = sorted(set(''.join(processed_lines)))  # English characters
urdu_characters = sorted(set(''.join(processed_lines_urdu)))  # Urdu characters

# Add start and end tokens to the vocab
english_characters = ['<start>', '<end>'] + english_characters
urdu_characters = ['<start>', '<end>'] + urdu_characters

# Create character-to-index mappings
english_char_index = {char: idx for idx, char in enumerate(english_characters)}
urdu_char_index = {char: idx for idx, char in enumerate(urdu_characters)}

# Reverse the character to index mappings for decoding
index_urdu_char = {i: char for char, i in urdu_char_index.items()}
index_english_char = {i: char for char, i in english_char_index.items()}


# Function to convert a sentence to one-hot encoding
def sentence_to_one_hot(sentence, char_index, max_length):
    one_hot = np.zeros((max_length, len(char_index)), dtype=np.float32)
    for i, char in enumerate(sentence):
        if char in char_index:  # Only map characters in the char_index
            if i < max_length:
                one_hot[i, char_index[char]] = 1
    return one_hot

# Find the maximum sentence lengths
max_english_length = max(len(sentence) for sentence in processed_lines) + 2  # Include <start> and <end> tokens
max_urdu_length = max(len(sentence) for sentence in processed_lines_urdu) + 2  # Include <start> and <end> tokens

# Prepare the encoder input data (for English sentences)
encoder_input_data = np.zeros((len(processed_lines), max_english_length, len(english_characters)), dtype=np.float32)
for i, sentence in enumerate(processed_lines):
    sentence = '<start> ' + sentence + ' <end>'  # Add start and end tokens
    encoder_input_data[i] = sentence_to_one_hot(sentence, english_char_index, max_english_length)

# Prepare the decoder input data (for Urdu sentences)
decoder_input_data = np.zeros((len(processed_lines_urdu), max_urdu_length, len(urdu_characters)), dtype=np.float32)
for i, sentence in enumerate(processed_lines_urdu):
    sentence = '<start> ' + sentence + ' <end>'  # Add start and end tokens
    decoder_input_data[i] = sentence_to_one_hot(sentence, urdu_char_index, max_urdu_length)

# Prepare the decoder target data (shifted by one timestep)
decoder_target_data = np.zeros_like(decoder_input_data)
decoder_target_data[:, :-1, :] = decoder_input_data[:, 1:, :]

# Output the arrays' shapes to verify
print("encoder_input_data shape:", encoder_input_data.shape)
print("decoder_input_data shape:", decoder_input_data.shape)
print("decoder_target_data shape:", decoder_target_data.shape)
```

```python
import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, Dense

#
# Define
latent_dim = 256  # You can adjust this value based on your data

# Define an input sequence and process it.
encoder_inputs = Input(shape=(None, num_encoder_tokens))
encoder = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
# We discard `encoder_outputs` and only keep the states.
encoder_states = [state_h, state_c]

# Set up the decoder, using `encoder_states` as initial state.
decoder_inputs = Input(shape=(None, num_decoder_tokens))
# We set up our decoder to return full output sequences,
# and to return internal states as well.
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs,
                                     initial_state=encoder_states)
decoder_dense = Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model that will turn encoder_input_data & decoder_input_data into decoder_target_data
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
          batch_size=64,
          epochs=10,
          validation_split=0.2)

# Save the trained model
model.save('translation_model.h5')

# Define the encoder model for inference (for translating new input)
```

```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense

latent_dim = 256  # Latent dimensionality of the encoding space

# Define an input sequence and process it.
encoder_inputs = Input(shape=(None, num_encoder_tokens))

# Add multiple LSTM layers to the encoder
encoder_lstm_1 = LSTM(latent_dim, return_sequences=True, return_state=True)
encoder_lstm_2 = LSTM(latent_dim, return_state=True)
encoder_outputs_1, _, _ = encoder_lstm_1(encoder_inputs)
encoder_outputs, state_h, state_c = encoder_lstm_2(encoder_outputs_1)
# Keep only the final states from the last layer
encoder_states = [state_h, state_c]

# Set up the decoder, using `encoder_states` as the initial state.
decoder_inputs = Input(shape=(None, num_decoder_tokens))

# Add multiple LSTM layers to the decoder
decoder_lstm_1 = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_lstm_2 = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs_1, _, _ = decoder_lstm_1(decoder_inputs, initial_state=encoder_states)
decoder_outputs, _, _ = decoder_lstm_2(decoder_outputs_1)

# Dense layer to generate the final output
decoder_dense = Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model that will turn
# `encoder_input_data` & `decoder_input_data` into `decoder_target_data`
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"])

# Summary of the model
model.summary()
```

```python
def decode_sequence(input_seq):
    # Encode the input as state vectors.
    states_value = encoder_model.predict(input_seq)  # Output states from all encoder LSTM layers

    # Generate empty target sequence of length 1.
    target_seq = np.zeros((1, 1, num_decoder_tokens))
    # Populate the first character of target sequence with the start character.
    target_seq[0, 0, target_token_index['\t']] = 1.

    # Sampling loop for a batch of sequences
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        # Pass the target sequence and current decoder states to the decoder model
        output_tokens, h1, c1, h2, c2 = decoder_model.predict(
            [target_seq] + states_value
        )

        # Sample a token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char

        # Exit condition: either hit max length or find stop character
        if (sampled_char == '\n' or len(decoded_sentence) > max_decoder_seq_length):
            stop_condition = True

        # Update the target sequence (of length 1)
        target_seq = np.zeros((1, 1, num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.

        # Update states (for all layers)
        states_value = [h1, c1, h2, c2]  # Update states for all LSTM layers

    return decoded_sentence
```

**Lstm with Attention mechanism**

```python
from keras.layers import Input, LSTM, Dense, AdditiveAttention, Concatenate
from keras.models import Model

# Encoder
encoder_inputs = Input(shape=(None, num_encoder_tokens))
encoder = LSTM(latent_dim, return_sequences=True, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
encoder_states = [state_h, state_c]  # Encoder final states

# Decoder
decoder_inputs = Input(shape=(None, num_decoder_tokens))
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)

# Attention mechanism
attention = AdditiveAttention(name="attention_layer")
context_vector = attention([decoder_outputs, encoder_outputs])  # Attention over encoder outputs

# Concatenate context vector with LSTM outputs
decoder_combined_context = Concatenate(axis=-1)([decoder_outputs, context_vector])

# Dense output layer
decoder_dense = Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_combined_context)

# Full model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"])

# Summary of the model
model.summary()
```

```python
def decode_sequence(input_seq):
    # Encode the input sequence to get encoder outputs and states
    encoder_outputs, state_h, state_c = encoder_model.predict(input_seq)
    states_value = [state_h, state_c]

    # Initialize the target sequence with the start character
    target_seq = np.zeros((1, 1, num_decoder_tokens))
    target_seq[0, 0, target_token_index['<start>']] = 1.

    # Decoding loop
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        # Pass the target sequence, encoder outputs, and states to the decoder
        output_tokens, h, c, attention_weights = decoder_model.predict(
            [target_seq] + states_value + [encoder_outputs])

        # Sample the next token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = reverse_target_char_index[sampled_token_index]
        decoded_sentence += sampled_char

        # Stop condition: end token or max length
        if (sampled_char == '<end>' or len(decoded_sentence) > max_decoder_seq_length):
            stop_condition = True

        # Update the target sequence
        target_seq = np.zeros((1, 1, num_decoder_tokens))
        target_seq[0, 0, sampled_token_index] = 1.

        # Update states
        states_value = [h, c]

    return decoded_sentence
```

```python
import tensorflow as tf
from tensorflow.keras import layers

class ScaledDotProductAttention(layers.Layer):
    def __init__(self, **kwargs):
        super(ScaledDotProductAttention, self).__init__(**kwargs)

    def call(self, query, key, value, mask=None):
        # Calculate the attention scores
        matmul_qk = tf.matmul(query, key, transpose_b=True)
        
        # Scale the attention scores by sqrt(dk)
        depth = tf.cast(tf.shape(key)[-1], tf.float32)
        scaled_attention_logits = matmul_qk / tf.sqrt(depth)

        # Apply the mask (optional) to the attention scores
        if mask is not None:
            scaled_attention_logits += (mask * -1e9)  # Mask with very large negative number
        
        # Compute the attention weights
        attention_weights = tf.nn.softmax(scaled_attention_logits, axis=-1)

        # Compute the context vector by multiplying attention weights with value
        output = tf.matmul(attention_weights, value)
        
        return output, attention_weights
```

```python
from tensorflow.keras import layers, Model
import tensorflow as tf

# Define the encoder
encoder_inputs = layers.Input(shape=(None, latent_dim))  # (batch_size, sequence_length, input_dim)
encoder_lstm = layers.LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_inputs)

# Pass the states to the decoder
encoder_states = [state_h, state_c]

# Define the decoder inputs
decoder_inputs = layers.Input(shape=(None, latent_dim))  # (batch_size, sequence_length, input_dim)

# Attention mechanism
attention = ScaledDotProductAttention()

# Query for the attention mechanism comes from the decoder state
decoder_lstm = layers.LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)

# Apply attention on the decoder outputs
context_vector, attention_weights = attention(decoder_outputs, encoder_outputs, encoder_outputs)

# Combine the context vector with decoder LSTM outputs
decoder_concat_input = layers.Concatenate(axis=-1)([context_vector, decoder_outputs])

# Output layer
decoder_dense = layers.Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_concat_input)

# Define the model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
import numpy as np
# Tokenize English and Urdu sentences
english_characters = sorted(set(''.join(processed_lines)))  # English characters
urdu_characters = sorted(set(''.join(processed_lines_urdu)))  # Urdu characters

# Add start and end tokens
english_characters = ['<start>', '<end>'] + english_characters
urdu_characters = ['<start>', '<end>'] + urdu_characters

# Create character-to-index mappings
english_char_index = {char: idx for idx, char in enumerate(english_characters)}
urdu_char_index = {char: idx for idx, char in enumerate(urdu_characters)}

# Reverse the mappings for decoding
index_urdu_char = {i: char for char, i in urdu_char_index.items()}
index_english_char = {i: char for char, i in english_char_index.items()}

# Function to convert a sentence to integer sequence
def sentence_to_integer_sequence(sentence, char_index, max_length):
    sequence = np.zeros(max_length, dtype=np.int32)  # Integer sequence with padding
    for i, char in enumerate(sentence):
        if char in char_index and i < max_length:
            sequence[i] = char_index[char]
    return sequence

# Find the maximum sentence lengths
max_english_length = max(len(sentence) for sentence in processed_lines) + 2  # Include <start> and <end>
max_urdu_length = max(len(sentence) for sentence in processed_lines_urdu) + 2  # Include <start> and <end>

# Prepare the encoder input data (English sentences as integer sequences)
encoder_input_data = np.zeros((len(processed_lines), max_english_length), dtype=np.int32)
for i, sentence in enumerate(processed_lines):
    sentence = '<start> ' + sentence + ' <end>'  # Add start and end tokens
    encoder_input_data[i] = sentence_to_integer_sequence(sentence, english_char_index, max_english_length)

# Prepare the decoder input data (Urdu sentences as integer sequences)
decoder_input_data = np.zeros((len(processed_lines_urdu), max_urdu_length), dtype=np.int32)
for i, sentence in enumerate(processed_lines_urdu):
    sentence = '<start> ' + sentence + ' <end>'  # Add start and end tokens
    decoder_input_data[i] = sentence_to_integer_sequence(sentence, urdu_char_index, max_urdu_length)

# Prepare the decoder target data (shifted by one timestep)
decoder_target_data = np.zeros_like(decoder_input_data)
decoder_target_data[:, :-1] = decoder_input_data[:, 1:]

# Output the shapes to verify
print("encoder_input_data shape:", encoder_input_data.shape)
print("decoder_input_data shape:", decoder_input_data.shape)
print("decoder_target_data shape:", decoder_target_data.shape)
```

```python
import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Embedding

# Define parameters
latent_dim = 256  # LSTM latent dimensionality
num_encoder_tokens = len(english_characters)  # Vocabulary size for English
num_decoder_tokens = len(urdu_characters)  # Vocabulary size for Urdu

# Define an input sequence and process it for the encoder
encoder_inputs = Input(shape=(None,), name="encoder_inputs")  # Integer sequences as input
encoder_embedding = Embedding(input_dim=num_encoder_tokens, output_dim=latent_dim, mask_zero=True, name="encoder_embedding")
encoder_embedded = encoder_embedding(encoder_inputs)

encoder = LSTM(latent_dim, return_state=True, name="encoder_lstm", use_cudnn=False)
encoder_outputs, state_h, state_c = encoder(encoder_embedded)
encoder_states = [state_h, state_c]

# Set up the decoder
decoder_inputs = Input(shape=(None,), name="decoder_inputs")  # Integer sequences as input
decoder_embedding = Embedding(input_dim=num_decoder_tokens, output_dim=latent_dim, mask_zero=True, name="decoder_embedding")
decoder_embedded = decoder_embedding(decoder_inputs)


decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True, name="decoder_lstm", use_cudnn=False)
decoder_outputs, _, _ = decoder_lstm(decoder_embedded, initial_state=encoder_states)

decoder_dense = Dense(num_decoder_tokens, activation='softmax', name="decoder_dense")
decoder_outputs = decoder_dense(decoder_outputs)

# Define the training model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit([encoder_input_data, decoder_input_data], decoder_target_data,
          batch_size=64,
          epochs=10,
          validation_split=0.2)

```

```python
from keras.models import Model
from keras.layers import Input

# Define the encoder model for inference
encoder_model = Model(encoder_inputs, encoder_states)

# Define the decoder model for inference
# Decoder state inputs (for maintaining context)
decoder_state_input_h = Input(shape=(latent_dim,), name="decoder_state_input_h")
decoder_state_input_c = Input(shape=(latent_dim,), name="decoder_state_input_c")
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

# Single timestep input for the decoder
decoder_inputs_single = Input(shape=(1,), name="decoder_inputs_single")  # One token at a time
decoder_embedded_single = decoder_embedding(decoder_inputs_single)  # Embed the input token

# Decoder LSTM step
decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_embedded_single, initial_state=decoder_states_inputs
)
decoder_states = [state_h, state_c]

# Final dense layer for generating probabilities over the vocabulary
decoder_outputs = decoder_dense(decoder_outputs)

# Define the complete decoder model
decoder_model = Model(
    [decoder_inputs_single] + decoder_states_inputs,  # Inputs: current token and states
    [decoder_outputs] + decoder_states               # Outputs: probabilities and updated states
)

# Save the models
encoder_model.save('encoder_model.h5')
decoder_model.save('decoder_model.h5')
```
